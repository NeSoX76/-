```python
# Встановлюємо TensorFlow і Keras Tuner, якщо вони ще не встановлені
!pip install tensorflow keras-tuner --quiet

# Після встановлення імпортуємо необхідні бібліотеки
import numpy as np
import matplotlib.pyplot as plt
from tensorflow.keras.datasets import fashion_mnist
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Flatten, Dropout
from tensorflow.keras.callbacks import EarlyStopping, ReduceLROnPlateau
from sklearn.metrics import classification_report
from keras_tuner import RandomSearch

# Завантажуємо датасет Fashion MNIST
(x_train, y_train), (x_test, y_test) = fashion_mnist.load_data()

# Нормалізація даних
x_train = x_train / 255.0
x_test = x_test / 255.0

# Виводимо перші 10 зображень
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_train[i], cmap='gray')
    plt.axis('off')
plt.show()

# Побудова моделі для завдання 1
def build_basic_model():
    model = Sequential([
        Flatten(input_shape=(28, 28)),
        Dense(128, activation='relu'),
        Dense(64, activation='relu'),
        Dense(10, activation='softmax')
    ])
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

model = build_basic_model()
model.summary()

# Рання зупинка та зменшення швидкості навчання
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)
reduce_lr = ReduceLROnPlateau(monitor='val_loss', factor=0.5, patience=3)

# Тренування моделі
history = model.fit(x_train, y_train, epochs=30, validation_data=(x_test, y_test),
                    callbacks=[early_stopping, reduce_lr])

# Збереження моделі у форматі .keras
model.save('/content/drive/MyDrive/fashion_mnist_basic_model.keras')

# Графіки функції втрат і точності
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['loss'], label='Training Loss')
plt.plot(history.history['val_loss'], label='Validation Loss')
plt.title('Функція втрат')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(history.history['accuracy'], label='Training Accuracy')
plt.plot(history.history['val_accuracy'], label='Validation Accuracy')
plt.title('Точність')
plt.legend()
plt.show()

# Класифікаційний звіт для завдання 1
y_pred = np.argmax(model.predict(x_test), axis=-1)
print("Класифікаційний звіт для базової моделі:")
print(classification_report(y_test, y_pred))

# Виведемо перші 10 зображень та їх передбачені й справжні класи
plt.figure(figsize=(10, 5))
for i in range(10):
    plt.subplot(2, 5, i + 1)
    plt.imshow(x_test[i], cmap='gray')
    plt.title(f"Прогноз: {y_pred[i]}\nСправжній: {y_test[i]}")
    plt.axis('off')
plt.show()

# Завдання 2: Налаштування гіперпараметрів за допомогою Keras Tuner
def build_model(hp):
    model = Sequential()
    model.add(Flatten(input_shape=(28, 28)))
    model.add(Dense(units=hp.Int('units', min_value=32, max_value=512, step=32), activation='relu'))
    model.add(Dropout(hp.Float('dropout', 0.1, 0.5, step=0.1)))
    model.add(Dense(10, activation='softmax'))
    
    model.compile(optimizer='adam', loss='sparse_categorical_crossentropy', metrics=['accuracy'])
    return model

# Використання Keras Tuner для підбору гіперпараметрів
tuner = RandomSearch(build_model, objective='val_accuracy', max_trials=5, executions_per_trial=3, directory='tuner_results')

# Пошук гіперпараметрів
tuner.search(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Найкраща модель
best_model = tuner.get_best_models(num_models=1)[0]
best_model.summary()

# Тренування найкращої моделі
best_history = best_model.fit(x_train, y_train, epochs=30, validation_data=(x_test, y_test),
                              callbacks=[early_stopping, reduce_lr])

# Збереження налаштованої моделі у форматі .keras
best_model.save('/content/drive/MyDrive/fashion_mnist_tuned_model.keras')

# Графіки функції втрат і точності для налаштованої моделі
plt.figure(figsize=(12, 5))
plt.subplot(1, 2, 1)
plt.plot(best_history.history['loss'], label='Training Loss (Tuned)')
plt.plot(best_history.history['val_loss'], label='Validation Loss (Tuned)')
plt.title('Функція втрат (налаштована модель)')
plt.legend()

plt.subplot(1, 2, 2)
plt.plot(best_history.history['accuracy'], label='Training Accuracy (Tuned)')
plt.plot(best_history.history['val_accuracy'], label='Validation Accuracy (Tuned)')
plt.title('Точність (налаштована модель)')
plt.legend()
plt.show()

# Класифікаційний звіт для налаштованої моделі
y_tuned_pred = np.argmax(best_model.predict(x_test), axis=-1)
print("Класифікаційний звіт для налаштованої моделі:")
print(classification_report(y_test, y_tuned_pred))

# Висновки: порівняння результатів
print("Порівняння базової моделі та моделі з налаштованими гіперпараметрами:")
print(f"Точність базової моделі: {max(history.history['val_accuracy'])}")
print(f"Точність налаштованої моделі: {max(best_history.history['val_accuracy'])}")
```
