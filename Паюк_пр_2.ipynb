```python
# Прізвище_пр_21.ipynb
import pandas as pd

# Завдання 1: Аналіз таблиці з Вікіпедії

# Завантаження даних з Вікіпедії
data_url = 'https://en.wikipedia.org/wiki/List_of_countries_by_GDP_(nominal)'
dfs = pd.read_html(data_url)

# Вибір таблиці з актуальними даними
df = dfs[2]

# Очищення та перетворення стовпців
df.columns = ['Country/Territory', 'IMF_Forecast', 'IMF_Year', 'WorldBank_Estimate', 'WorldBank_Year', 'UN_Estimate', 'UN_Year']
df = df.drop(columns=['IMF_Year', 'WorldBank_Year', 'UN_Year'])

# Перетворення стовпців до числових значень
df['IMF_Forecast'] = pd.to_numeric(df['IMF_Forecast'].str.replace(',', ''), errors='coerce')
df['WorldBank_Estimate'] = pd.to_numeric(df['WorldBank_Estimate'].str.replace(',', ''), errors='coerce')
df['UN_Estimate'] = pd.to_numeric(df['UN_Estimate'].str.replace(',', ''), errors='coerce')

# Перевірка типів даних
print("\nЗавдання 1: Типи даних:")
print(df.dtypes)

# Перевірка на пропущені значення
print("\nЗавдання 1: Перевірка на пропущені значення:")
print(df.isna().sum())

# Кількість дублікатів
print("\nЗавдання 1: Кількість дублікатів:")
print(df.duplicated().sum())

# Описова статистика
print("\nЗавдання 1: Описова статистика:")
print(df.describe(include='all'))

# Країни з найбільшим відхиленням між IMF_Forecast та WorldBank_Estimate
df['Difference'] = df['IMF_Forecast'] - df['WorldBank_Estimate']
print("\nЗавдання 1: Країни з найбільшим відхиленням між IMF_Forecast та WorldBank_Estimate:")
print(df[['Country/Territory', 'IMF_Forecast', 'WorldBank_Estimate', 'Difference']].sort_values(by='Difference', ascending=False).head())

# Кореляція між показниками
print("\nЗавдання 1: Кореляція між показниками:")
print(df[['IMF_Forecast', 'WorldBank_Estimate', 'UN_Estimate']].corr())

# Середнє значення для кожного стовпця
print("\nЗавдання 1: Середнє значення для кожного стовпця:")
print(df[['IMF_Forecast', 'WorldBank_Estimate', 'UN_Estimate']].mean())

# Країна з найвищою варіативністю у показниках між роками
df['Std_Dev'] = df[['IMF_Forecast', 'WorldBank_Estimate', 'UN_Estimate']].std(axis=1)
print("\nЗавдання 1: Країна з найвищою варіативністю у показниках між роками:")
print(df[['Country/Territory', 'Std_Dev']].sort_values(by='Std_Dev', ascending=False).head())

# Висновки для завдання 1
print("\nВисновки для Завдання 1:")
print("1. Найбільше відхилення між прогнозами IMF і оцінками Світового банку спостерігається у 'World', США і Китаї.")
print("2. Всі показники мають високу кореляцію, що свідчить про узгодженість оцінок між різними джерелами.")
print("3. Середні значення для кожного показника надають уявлення про масштаб економік.")
print("4. Країна з найбільшою варіативністю показників між роками - 'World', що вказує на великі коливання у глобальних оцінках.")

# Завдання 2: Аналіз даних про Титанік

# Завантаження даних про Титанік з іншого джерела
titanic_url = 'https://raw.githubusercontent.com/datasciencedojo/datasets/master/titanic.csv'
titanic_df = pd.read_csv(titanic_url)

# Виведення перших кількох рядків таблиці
print("\nЗавдання 2: Перші кілька рядків таблиці:")
print(titanic_df.head())

# Перевірка типів даних
print("\nЗавдання 2: Типи даних:")
print(titanic_df.dtypes)

# Перевірка на пропущені значення
print("\nЗавдання 2: Перевірка на пропущені значення:")
print(titanic_df.isna().sum())

# Кількість дублікатів
print("\nЗавдання 2: Кількість дублікатів:")
print(titanic_df.duplicated().sum())

# Описова статистика
print("\nЗавдання 2: Описова статистика:")
print(titanic_df.describe(include='all'))

# Висновки для завдання 2
print("\nВисновки для Завдання 2:")
print("1. Дані про Титанік містять численні пропущені значення, зокрема у стовпцях 'Age', 'Cabin' та 'Embarked'.")
print("2. Типи даних у таблиці варіюються, включаючи числові та категоріальні значення.")
print("3. Описова статистика допомагає зрозуміти розподіл даних про пасажирів, включаючи їхній вік, клас та виживання.")
print("4. Відсутність дублікатів у даних підтверджує унікальність записів.")

```
